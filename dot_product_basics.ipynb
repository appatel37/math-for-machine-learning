# Dot Product – Linear Algebra for Machine Learning

This notebook explains the dot product operation and why it is fundamental in machine learning.

## What is Dot Product?

The dot product is an operation that takes two vectors of the same length
and returns a single number.

It is calculated by multiplying corresponding elements and adding them together.

If we have two vectors:

x = [x1, x2, x3]  
w = [w1, w2, w3]

Dot product = x1*w1 + x2*w2 + x3*w3

```python
import numpy as np

x = np.array([1, 2, 3])
w = np.array([4, 5, 6])

np.dot(x, w)
```

## Dot Product in Machine Learning

In machine learning, predictions are often calculated as:

y = X · w + b

Where:
- X is the feature matrix
- w is the weight vector
- b is the bias (scalar)

The dot product combines features with weights.

```python
# one data point with 3 features
features = np.array([3500, 3, 2])

# weights learned by the model
weights = np.array([120, 15000, 20000])

prediction = np.dot(features, weights)
prediction
```

## My Understanding

The dot product multiplies features with their importance (weights)
and adds them together to produce a prediction.
